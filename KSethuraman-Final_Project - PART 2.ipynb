### Cleaning the Text and Identifying the Keywords of an Article 

```python
#Cleaning the Headline of all the Articles
import ast
for i in range(0, len(art)):
    art['headline'][i]=ast.literal_eval(art['headline'][i])
    art['headline'][i]=art['headline'][i]['print_headline']
art['headline'][0]
```

```python
#Package to get allthe Stop Words. Run the Download Step Only if RUnning for the 1st Time

#nltk.download('stopwords')  
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
```

```python
#Getting all the Stop Words

stop_words = stopwords.words('english')
stop_words[:10]
stop_words+=['mr','mr.']
```

```python
# Function to Strip all ALpha Numeric Characters from a Text

def stripNonAlphaNum(text):
    import re
    return re.compile(r'\W+', re.UNICODE).split(text)
```

```python
#Tokenising the Article Content and Removing Non Alpha Numeric Characters

art['words']=''
for i in range(0, len(art)):
    if art['article'][i] is not None:
        art['words'][i]=stripNonAlphaNum(art['article'][i])
        art['words'][i]=art['words'][i]

        #art['words'][i]=art['article'][i].split()

art['words'][2]
```

```python
#Removing all Punctuations and getting the non-stop Words alone

table = str.maketrans('', '', string.punctuation)
from string import punctuation
import re
art['non_stop_words']=''


for i in range(0, len(art)):
    wordsFiltered = []
    for word in art['words'][i]: 
        word = word.lower().rstrip(punctuation)
        word=word.translate(table)
        if word not in stop_words:
            if len(word)>2:
                #print(word)
                wordsFiltered.append(word)
    art['non_stop_words'][i] =wordsFiltered
art['non_stop_words'][1]
```

```python
#A Function to get the frequency of all the Non-Stop WOrds in Each article

def word_count(str):
    counts = dict()

    for word in str:
        if word in counts:
            counts[word] += 1
        else:
            counts[word] = 1
    return counts

# TO Show how the Function WOrks
print( word_count(art['non_stop_words'][0]))
```

```python
#GETTING THE 10 MOST FREQUENT WORDS FROM EACH ARTICLE

art['top_freq']=''

for i in range(0,len(art)):
    art['top_freq'][i]=Counter(art['non_stop_words'][i]).most_common(10)
    
art['top_freq'][1]
```

```python
#GETTING THE 10 MOST FREQUENT WORDS AS A LIST OF WORDS WITHOUT FREQUENCY

from heapq import nlargest
art['top_words']=''


for i in range(0,len(art)):
    d=dict()
    d.update(word_count(art['non_stop_words'][i]))
    Highest = nlargest(10, d, key = d.get)
    art['top_words'][i]=Highest

art['top_words_n']=''

for i in range(0,len(art)):
    art['top_words_n'][i]=" ".join(str(x) for x in art['top_words'][i])

#TO DEMONSTRATE HOW IT WORKS
art['top_words'][0]
```

```python
#CREATING AN ID COLUMN 

art['ID'] = np.arange(len(art))

#FILTERING ONLY FOR REUQUIRED ARTICLES
art_n = art[['ID','headline','web_url','article','top_words_n']]
art_n.head()
```

# Part 4: Assesing the Similarity between the Documents to Filter for the Most Relevant ones    -     Using TF -IDF

#### In order to Find Similarity We are getting the TF-IDF of Keywords in all the articles

```python
# Grouping the Data by ID and Splitting the words in to Columns
rows = list()

for row in art[['ID', 'top_words']].iterrows():
    r = row[1]
    for word in r.top_words:
        rows.append((r.ID, word))
    

key_words = pd.DataFrame(rows, columns=['ID', 'top_words'])
key_words = key_words[key_words.top_words.str.len() > 0]

key_words.head()
```

```python
#Computing the Term Frequency of each Key word - ID Wise
counts = key_words.groupby('ID')\
    .top_words.value_counts()\
    .to_frame()\
    .rename(columns={'top_words':'n_w'})
counts.head()
key_word_sum = counts.groupby(level=0)\
    .sum()\
    .rename(columns={'n_w': 'n_d'})
key_word_sum.head()
```

#### Getting the TF - Term Frequency

```python
tf = counts.join(key_word_sum)
tf['tf'] = tf.n_w/tf.n_d
tf.head()
```

#### Getting the IDF - Inverse Document Frequency

```python
# Getting the IDF and joining it to the Data Frame

idf = key_words.groupby('top_words')\
    .ID\
    .nunique()\
    .to_frame()\
    .rename(columns={'ID':'i_d'})\
    .sort_values('i_d')
idf.head()

c_d = key_words.top_words.nunique()
idf['idf'] = np.log(c_d/idf.i_d.values)
idf.head()

tf_idf = tf.join(idf)
tf_idf['tf_idf'] = tf_idf.tf * tf_idf.idf
tf_idf.head()
```

```python
# Grouping the Data based on ID and Sorting it based on DESCENDING TF-IDF

tf_idf=tf_idf.sort_values(['ID','tf_idf'],ascending=False).groupby('ID').head(5)
tf_idf.head()
```
